```python
#cleaning & structuring our data
import json
```

```python
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data)
print(type(data))
```

```python
#cleaning & structure the data
def clean_data(data):
    text_to_num = {"one":1,"two":2,"three":3,"four":4,"five":5}

    cleaned_data= []
    unique_users = set()
    for user in data:
        #clean rating
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating
        #handle missing data
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"]= None
        #duplicate data

        if(user["name"].strip() in unique_users):
            continue
        unique_users.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
        
        
```

```python
cleaned_data = clean_data(data)
for data in cleaned_data:
    print(data)
```

```python
#assignment 2 
#get meaninnful insights from data
def get_insights(data):
    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"avg rating = {tot_rating/len(data)}")
    #percentage of users with poor rating
    poor_rating = 0
    for user in data:
        if(float(user["rating"])<3):
            poor_rating+= 1
    print(f"% of poor rating = {poor_rating/len(data)*100}")
```

```python
get_insights(cleaned_data)
```

```python
#assignment 3  
#if user rating >=4 same brand product
#if user rating <4 different brand product
def get_recommendation(data):
    recommendation = []
    
    for user in data:
        curr_recommendation = {}
        curr_recommendation["name"] = user["name"]
        if(float(user["rating"]) >=4):
            curr_recommendation["brand"] = "apple"
        else:
            curr_recommendation["brand"] = "samsung"
        recommendation.append(curr_recommendation)
    return recommendation    
```

```python
get_recommendation(cleaned_data)
```
